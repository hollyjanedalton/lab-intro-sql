```python
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
engine = create_engine(connection_string)
#this creates a connection to the database
#now after this, no longer accessing the database in SQL directly, 
#just working on local copy in jupyter. not making any changes to the SQL database
```

```python
#2 How many distinct (different) actors' last names are there?
query = '''SELECT COUNT(DISTINCT(last_name)) 
AS 'Count of distinct last names' 
FROM sakila.actor'''
print(query)
data = pd.read_sql_query(query, engine)
data
```

```python
#3 Add an additional column day_type with values 'weekend' and 'workday'
#depending on the rental day of the week.
#this was a problem in one of the first SQL labs

#rental day of the week.

query = '''SELECT *, date_format(rental_date, '%%M') 
AS 'month_of_rental', 
date_format(rental_date, '%%W') 
AS 'weekday_of_rental',
date_format(return_date, '%%M') 
AS 'month of return',
date_format(return_date, '%%W') 
AS 'day_of_return',
CASE 
WHEN (date_format(return_date, '%%W') = 'Saturday') or (date_format(return_date, '%%W') = 'Sunday')
THEN 'weekend'
ELSE 'workday'
END AS day_type
FROM sakila.rental'''
data = pd.read_sql_query(query, engine)
data
```

```python
#4 Get all films with ARMAGEDDON in the title.
#in python need to add 2 %% instead of one
query = "SELECT title FROM sakila.film WHERE title LIKE ('%%ARMAGEDDON%%')"
data = pd.read_sql_query(query, engine)
data
```

```python
#5 Get 10 the longest films.
query = '''SELECT title as movie_title FROM sakila.film
ORDER BY length DESC
LIMIT 10'''
data = pd.read_sql_query(query, engine)
data
```

```python
#6 How many films include Behind the Scenes content?
query = '''SELECT COUNT(special_features) as films_with_behind_the_scenes_content
FROM sakila.film
WHERE special_features like ('%%Behind the Scenes%%')'''
data = pd.read_sql_query(query, engine)
data
```

```python
#7 Which kind of movies (rating) have a mean duration of more than two hours?


query = '''SELECT rating, ROUND(AVG(length),2) as "avg length in minutes"
FROM sakila.film
GROUP BY rating
HAVING ROUND(AVG(length),2) > 120'''
data = pd.read_sql_query(query, engine)
data
```

```python
#8 Rank films by length (filter out the rows that have nulls or 0s in length column). 
#In your output, only select the columns title, length, and the rank.

query = '''SELECT title as film_title, length as length_in_minutes
FROM sakila.film
WHERE length <> 0 OR null
ORDER BY length ASC;'''
data = pd.read_sql_query(query,engine)
data
```
